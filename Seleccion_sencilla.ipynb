```python
"""
El número de sonidos emitidos por un grillo en un minuto es una función de la temperatura. Por lo tanto, es posible determinar el nivel de la temperatura haciendo uso de un
grillo como termómetro. La fórmula para la función es : T = N/4 + 40 Donde T representa la temperatura en grados Fahrenheit y N el número de sonidos emitidos por minuto.
Calcule e imprima el valor de T para cualquier valor positivo de N.
"""

sonidos = int(input("ingresa el num de sonidos x minuto: "))
temp = (sonidos/4) + 40
print("La temperatura es: ", temp)
```

```python
"""
Dados los valores enteros P y Q, determine si los mismos satisfacen la siguiente expresión: P^3 + Q^4 - 2P^2 < 680 . En caso afirmativo debe imprimir los valores P y Q.
"""

p = int(input("Ingresa el valor de p: "))
q = int(input("Ingresa el valor de q: "))
expresion = p**3 + q**4 - 2*p**2
if(expresion < 680):
    print("p = ", p)
    print("q = ", q)
```

```python
"""
Calcular e imprimir las raices reales de las ecuaciones dadas siemore y cuando b**2 - ac de un valor positivo
"""
a = int(input("Ingrese el valor de a: "))
b = int(input("Ingrese el valor de b: "))
c = int(input("Ingrese el valor de c: "))
determinante =  b**2 - 4*a*c
if determinante > 0:
    x1 = (0 -(b) + math.sqrt(determinante))/(2*a)
    x2 = (0 -(b) - math.sqrt(determinante))/(2*a)
    print("El valor de X1 es: ", x1)
    print("El valor de X2 es: ", x2)
else:
  print("El determinante es negativo o 0")
```

```python
"""
En la escuela se paga un costo de inscripción de $2500 pesos. A los alumnos con promedio mayor o igual a 90 se les da un descuento del 15%. Con base en el promedio del alumno,
calcule e imprima cuánto deberá pagar de inscripción.
"""
inscripcionBase = 2500
prom = float(input("Ingrese el promedio del alumno: "))
if prom >= 90:
    descuento = inscripcionBase * 0.15
    inscripcion = inscripcionBase - descuento
else:
    inscripcion = inscripcionBase
print(f"El costo de inscripción a pagar es: ${inscripcion}")
```

```python
"""
Una empresa paga a sus empleados el salario mínimo en las primeras 40 horas de trabajo, y paga salario mínimo y medio (1 1⁄2 salario mínimo) por todas las horas que excedan de
las 40. Dados el salario mínimo y la cantidad de horas trabajadas, calcule e imprima el sueldo del trabajador.
"""
horasMinimo = 40
salarioMinimo = float(input("Ingrese el salario mínimo por hora: "))
horasTrabajadas = float(input("Ingrese la cantidad de horas trabajadas: "))

if horasTrabajadas > horasMinimo:
    horasExtras = horasTrabajadas - horasMinimo
    salarioInicial = horasMinimo * salarioMinimo
    salarioExtra = horasExtras * (1.5 * salarioMinimo)
    sueldo = salarioInicial + salarioExtra
else:
    sueldo = salarioMinimo * horasTrabajadas
print(f"El sueldo total del trabajador es: ${sueldo}")
```
