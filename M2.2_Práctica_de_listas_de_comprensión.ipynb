```python
# Escribe una lista de comprensión que genere la lista de todos los números entre 1 y 100 que son divisibles en 8.

divisibles = []
for num in range(9,101):
  if num % 8 == 0:
    divisibles.append(num)
print(divisibles)
```

```python
# Escribe una lista de comprensión que genere la lista de todos los números entre 1 y 100 que contengan un 6.

lista = []
for num in range(1,101):
  if '6' in str(num):
    lista.append(num)
print(lista)
```

```python
# Pide al usuario una cadena y escribe una lista de comprensión que devuelva la cadena sin espacios.

cadena = str(input("Ingresa una cadena: "))
noEspacios = cadena.replace(" ", "")
print(noEspacios)

# Escribe una lista de comprensión que devuelva una cadena sin las vocales (utiliza la cadena del ejercicio anterior).
noVocales = []
for i in cadena:
  if i.lower() not in 'aeiou':
        noVocales += i

s = ''.join(noVocales)
print(s)
```

```python
# Pide el nombre de un archivo, léelo línea por línea y devuelve todas las palabras que tienen 5 letras o menos. (Utiliza el archivo words.txt para probar).

archivo = input("Ingrese el nombre de un archivo para leer: ")
try:
    arc = open(archivo, 'r')
    A = []
    for n in arc:
        palabra = n.split()
        for i in palabra:
            if len(i) <= 5:
                A.append(i)
    print(A)
except:
    print("El archivo no se encuentra")
```

```python
# Escribe una lista de comprensión que genere la lista de todos los números entre 1 y 100 que son divisibles por cada dígito además de 1 (2-9)

A = []
for primo in range(1,101):
    n = 0
    for i in range(1,10):
        if primo % i == 0:
            n = n + 1
    if n == 1:
        A.append(primo)
print(A)
```

```python
# Escribe una lista de comprensión que genere una lista de todos los números entre 1 y 100 del máximo divisor de cada número.

A = []
max = 0
for n in range(1,101):
    for i in range(1,n):
        if n%i == 0:
            max = i
    A.append([n, max])
print(A)
```
